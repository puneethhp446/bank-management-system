```python
  # BANK MANAGEMENT SYSTEM DATABASE
accounts = []

# Function to create account
def create_account(acc_no, name, deposit, acc_type):
    # Check if account number already exists
    for acc in accounts:
        if acc["acc_no"] == acc_no:
            print("Account Number Already Exists! Please enter a new number.")
            return
    
    # Check minimum deposit requirement
    if deposit < 500:
        print("Initial deposit must be at least 500!")
        return
    
    account = {
        "acc_no": acc_no,
        "name": name,
        "balance": float(deposit),
        "type": acc_type
    }
    accounts.append(account)
    print("Account Created Successfully!")

# Function to deposit money
def deposit(acc_no, amount):
    for acc in accounts:
        if acc["acc_no"] == acc_no:
            acc["balance"] += amount
            print("Deposit Successful!")
            return
    print("Account Not Found!")

# Function to withdraw money
def withdraw(acc_no, amount):
    for acc in accounts:
        if acc["acc_no"] == acc_no:
            if acc["balance"] - amount >= 500:  # Maintain minimum balance
                acc["balance"] -= amount
                print("Withdrawal Successful!")
            else:
                print(f"Insufficient Balance! You must maintain a minimum of 500. "
                      f"You can withdraw up to {acc['balance'] - 500}.")
            return
    print("Account Not Found!")

# Function to check balance
def check_balance(acc_no):
    for acc in accounts:
        if acc["acc_no"] == acc_no:
            print(f"Account Holder: {acc['name']}")
            print(f"Account Type: {acc['type']}")
            print(f"Balance: {acc['balance']}")
            return
    print("Account Not Found!")

# Function to calculate interest (only for savings accounts)
def calculate_interest(acc_no):
    for acc in accounts:
        if acc["acc_no"] == acc_no:
            if acc["type"].lower() == "savings":
                interest = acc["balance"] * 0.0725
                print(f"Yearly Interest for {acc['name']} (Savings): {interest}")
            else:
                print("Interest calculation is only for Savings accounts.")
            return
    print("Account Not Found!")

# Function to display all accounts
def display_all():
    print("All Accounts:")
    for acc in accounts:
        acc_tuple = (acc["acc_no"], acc["name"], acc["balance"], acc["type"])
        print(f"Acc No: {acc_tuple[0]}, Name: {acc_tuple[1]}, "
              f"Balance: {acc_tuple[2]}, Type: {acc_tuple[3]}")

# Main program loop
while True:
    print("\n---- Bank Management System ----")
    print("1. Create Account")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Check Balance")
    print("5. Display All Accounts")
    print("6. Calculate Interest (Savings)")
    print("7. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        acc_no = input("Enter Account Number: ")
        name = input("Enter Name: ")
        deposit_amt = float(input("Enter Initial Deposit: "))
        acc_type = input("Enter Account Type (Savings/Current): ")
        create_account(acc_no, name, deposit_amt, acc_type)

    elif choice == 2:
        acc_no = input("Enter Account Number: ")
        amount = float(input("Enter Amount to Deposit: "))
        deposit(acc_no, amount)

    elif choice == 3:
        acc_no = input("Enter Account Number: ")
        amount = float(input("Enter Amount to Withdraw: "))
        withdraw(acc_no, amount)

    elif choice == 4:
        acc_no = input("Enter Account Number: ")
        check_balance(acc_no)

    elif choice == 5:
        display_all()

    elif choice == 6:
        acc_no = input("Enter Account Number: ")
        calculate_interest(acc_no)

    elif choice == 7:
        print("Exiting... Thank you!")
        break

    else:
        print("Invalid Choice! Try again.")
```
